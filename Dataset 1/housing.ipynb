```python
import pandas as pd

# Data
data = [1, 2, 3, 4, 5, 5, 5]

# Create DataFrame
df = pd.DataFrame(data, columns=["values"])

# Mean (Average)
mean_value = df["values"].mean()

# Median
median_value = df["values"].median()

# Mode
mode_value = df["values"].mode()

print("Mean (Average):", mean_value)
print("Median:", median_value)
print("Mode:")
print(mode_value)
```

```python
plt.figure(figsize=(8,5))
plt.hist(dataset["bedrooms"],bins = 30)

plt.title("house price Distribution")
plt.xlabel("price")
plt.ylabel("Frequency")
plt.show()
```

```python
# IMPortant 
#हा code डेटासेटमधील सर्व numeric columns चा correlation heatmap दाखवतो, numbers annotate करून आणि figure size set करून.
#This code shows a correlation heatmap of all numeric columns in the dataset with numbers annotated and a custom figure size.
plt.figsize=(8,5)
sns.heatmap(dataset.corr(numeric_only=True), annot=True, fmt=".2f")
plt.title("correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(dataset["area"], dataset["price"])
plt.xlabel("Area (sqft)")
plt.ylabel("Price")
plt.title("Area vs price")
plt.show()
```
