```python
#Numeric Data Types
#1. Integer
#Integer stores whole numbers without decimal point.
#Integer मध्ये पूर्ण संख्या साठवतात. दशांश नसतो.

a = 10
b = 12
```

```python
# 2.Float
#Float stores numbers with decimal point.
#Float मध्ये दशांश असलेल्या संख्या साठवतात.

b = 10.5

```

```python
# Complex number
#Complex number has real and imaginary part. Imaginary part uses j.
#Complex number मध्ये real आणि imaginary भाग असतो. Imaginary साठी j वापरतात.

c = 2 + 3j
print(c)
```

```python
#Dictionary
#Dictionary stores data in key-value pairs.
#Dictionary मध्ये data key आणि value च्या जोडीने साठवला जातो.

student = {"name": "damu", "age": 20}
```

```python
#Boolean has only two values: True or False.
#Boolean मध्ये फक्त दोन values असतात: True किंवा False.

result = True

print(10 > 9)
print(10 == 9)
print(10 < 9)



a = 200
b = 33

if b > a:
  print("b is greater than a")
else:
  print("b is not greater than a")



#Evaluate a string and a number:
print(bool("Hello"))
print(bool(15))

#Evaluate two variables:
x = "Hello"
y = 15

print(bool(x))
print(bool(y))
```

```python
#set
#Set stores unique values (no duplicates).
#Set मध्ये एकसारख्या values परत परत येत नाहीत.

s = {1, 2, 3}
print(s)


fruites = {"apple", "banana", "cherry"}
print(fruites)


# From a list (removes duplicates)
numbers_list = [1, 2, 2, 3, 4, 3, 5]
unique_numbers = set(numbers_list)
print(unique_numbers) 

# Creating an empty set ({} creates an empty dictionary)
empty_set = set()
print(type(empty_set)) 


# Input
#input() is used to take input from the user through the keyboard.
#input() वापरकर्त्याकडून (keyboard वरून) माहिती घेण्यासाठी वापरतात.

name = input("Enter your name: ")
print(name)

age = int(input("Enter age: "))
print(age)

price = float(input("Enter price: "))
print(price)


a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print(a + b)


```

```python
#sequence Data Types
#Sequence data types store data in ordered form.
#Sequence data types मध्ये data ठराविक क्रमाने साठवला जातो.

#1.String(str)
name = "Python"
print(name)

#2.list
#List stores multiple values and can be changed.
#List मध्ये अनेक values साठवता येतात आणि त्या बदलता येतात.

marks = [70, 80, 90]
print(marks)


x = ["geeks", "for", "geeks"]
print(x[1])
x[1]="hello"
print(x)




#3. Tuple
#Tuple stores multiple values but cannot be changed.
#Tuple मध्ये अनेक values साठवतात पण त्या बदलता येत नाहीत.

t = (10, 20, 30)
print(t)


```
